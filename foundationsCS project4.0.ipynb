## 17.Build chains of trips. Two trips are consecutive in a chain if (a) they have the same VendorID, (b) the pickup location of the second trip is also the dropoff location of the first trip, (c) the pickup time of the second trip is after the dropoff time of the first trip, and (d) the pickup time of the second trip is at most 2 minutes later than the dropoff time of the first trip.

# Codice della Funzione

```python
def identify_trip_chains1(df):
    
    df = df.sort_values(['VendorID', 'tpep_pickup_datetime']).reset_index(drop=True)
    df['chain'] = -1 

    time_threshold = pd.Timedelta(minutes=2)
    current_chain = 0

    
    for vendor_id, vendor_df in df.groupby('VendorID', sort=False):
        indices = vendor_df.index.values
        n = len(indices)
        pickup_times = vendor_df['tpep_pickup_datetime'].values
        dropoff_times = vendor_df['tpep_dropoff_datetime'].values
        PULocationIDs = vendor_df['PULocationID'].values
        DOLocationIDs = vendor_df['DOLocationID'].values
        chains = np.full(n, -1, dtype=int)

        active_chains = defaultdict(deque)
        chain_locations = {}

        for i in range(n):
            current_pickup = pickup_times[i]
            current_dropoff = dropoff_times[i]
            current_PULocationID = PULocationIDs[i]
            current_DOLocationID = DOLocationIDs[i]

            chain_list = active_chains[current_PULocationID]
            while chain_list and (current_pickup - chain_list[0][0]) > time_threshold:
                _, chain_id = chain_list.popleft()
                del chain_locations[chain_id]

            chain_found = False
            for idx, (dropoff_time, chain_id) in enumerate(chain_list):
                time_diff = current_pickup - dropoff_time
                if pd.Timedelta(0) < time_diff <= time_threshold:
                    chains[i] = chain_id
                    chain_found = True

                    chain_list.remove((dropoff_time, chain_id))
                    active_chains[current_DOLocationID].append((current_dropoff, chain_id))
                    chain_locations[chain_id] = (current_dropoff, current_DOLocationID)
                    break

            if not chain_found:
                current_chain += 1
                chains[i] = current_chain
                active_chains[current_DOLocationID].append((current_dropoff, current_chain))
                chain_locations[current_chain] = (current_dropoff, current_DOLocationID)

        df.loc[indices, 'chain'] = chains

    return df
```

# COMMENTO

```python
def identify_trip_chains1(df):
    
    # Trova delle chain di viaggi a seconda che verifichino i seguenti vincoli:
    # a) il venditore è lo stesso 
    # b) la pickup location del secondo viaggio = drop off location del viaggio precedente
    # c) il pickup time del secondo viaggio deve essere successivo al drop off time del primo
    # d) il pickup time del secondo deve esssere max 2 minuti dopo il drop off time del primo
    
    # La funzione accetta come argomento il dataframe e restituisce il dataframe con l'aggiunta
    # della colonna chain
    
    
    # Prima si ordina il dataset per Vendor e Pickup time
    df = df.sort_values(['VendorID', 'tpep_pickup_datetime']).reset_index(drop=True)
    df['chain'] = -1  # Creazione nuova colonna 'chain' (il -1 indica che l'osservazione non appartiene a nessuna chain)

    time_threshold = pd.Timedelta(minutes=2) # la differenza di tempo massima tra i due trip 
    current_chain = 0 # inizio contatore per assegnare le chain

    # Si raggruppa per venditore (ciclo prima con tutti i vendor1 e poi tutti i vendor1, così da non 
    # controllare per i vendor 2 anche tutte le catene iniziate dai vendor1)
    for vendor_id, vendor_df in df.groupby('VendorID', sort=False): # sort = False perchè abbiamo già ordinato prima
        
        # Estraiamo le metriche necessarie per confrontare i viaggi come array tramite NumPy per ottimizzare i check
        indices = vendor_df.index.values
        n = len(indices)
        pickup_times = vendor_df['tpep_pickup_datetime'].values
        dropoff_times = vendor_df['tpep_dropoff_datetime'].values
        PULocationIDs = vendor_df['PULocationID'].values
        DOLocationIDs = vendor_df['DOLocationID'].values
        chains = np.full(n, -1, dtype=int) # array per memorizzare le chain (come la colonna chain, -1 indica che 
                                           # il trip non è stato assegnato ad una chain

        # Per ogni venditore, crea un dizionario dove le chiavi sono le Pickup location e i valori sono 'deque objects'
        # con dentro le info delle chain attive in quella location. (in pratica permette di controllare solo le chain con
        # la drop off location uguale alla pickup location)
        active_chains = defaultdict(deque)
        
        # Si crea un dizionario che ha come chiavi il numero della chain, e come valori le info sull'ultimo trip di 
        # quella chain. (in pratica, essendo i trip ordinati per pickup time, permetterà di eliminare le chain da 
        # prendere in considerazione se risultano più 'vecchie' di 2 minuti)
        chain_locations = {}

        # Per ogni trip, vengono estratte le metriche necessarie per confrontare l'i-esimo trip e le chain presenti in
        # active_chains
        for i in range(n):
            current_pickup = pickup_times[i]
            current_dropoff = dropoff_times[i]
            current_PULocationID = PULocationIDs[i]
            current_DOLocationID = DOLocationIDs[i]

            # Si comincia filtrando le chain che hanno come chiave (cioè drop off location) lo stesso valore della
            # pickup location del trip corrente (in questo modo si evita di controllare chain inutili)
            chain_list = active_chains[current_PULocationID]
            # Si continua rimuovendo le chain che hanno un drop off time troppo 'vecchio'
            while chain_list and (current_pickup - chain_list[0][0]) > time_threshold:
                # Estrae le chain_id delle chain 'vecchie' (.popleft() è fatto apposta per gli oggetti deques)
                _, chain_id = chain_list.popleft()
                # Si rimuove la chain dal dizionario
                del chain_locations[chain_id]

            # Si prova a cercare una chain valida per il trip corrente
            chain_found = False 
            # Si itera per tutte le chain recenti (time_diff <= minuti) con dropoff location = pickup location 
            # del trip corrente (enumerate prende anche l'index di chain_list che è la chain_id)
            for idx, (dropoff_time, chain_id) in enumerate(chain_list):
                time_diff = current_pickup - dropoff_time
                if pd.Timedelta(0) < time_diff <= time_diff:
                    # Se c'è un match allora si salva la chain_id
                    chains[i] = chain_id
                    chain_found = True

                    # Si rimuove la vecchia per aggiornare la chain
                    chain_list.remove((dropoff_time, chain_id))
                    # Si aggiorna il nuovo drop off time e la drop off location
                    active_chains[current_DOLocationID].append((current_dropoff, chain_id))
                    chain_locations[chain_id] = (current_dropoff, current_DOLocationID)
                    # Si chiude il loop appena aggiorna la chain
                    break  
            # Se non trova una chain, il trip corrente ne inizia una nuova
            if not chain_found:
                # Aggiorna il contatore per la chain_id
                current_chain += 1
                chains[i] = current_chain
                # Aggiunge la chain e le metriche per i confronti nei due dizionari
                active_chains[current_DOLocationID].append((current_dropoff, current_chain))
                chain_locations[current_chain] = (current_dropoff, current_DOLocationID)

        # L'array con le chain_id viene inserito nella colonna 'chain' del dataframe di partenza
        df.loc[indices, 'chain'] = chains

    return df
```
