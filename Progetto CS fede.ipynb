## 11.For each payment type, determine the interval when the overall ratio between the tip and the fare amounts is maximum

```python
df['overall_ratio']=df['tip_amount']/df['fare_amount']
```

```python
overall_ratio = df.groupby(['payment_type', 'clustered_pu_time'])['overall_ratio'].mean().reset_index()
overall_ratio = overall_ratio[overall_ratio['overall_ratio'].notnull()]
overall_ratio.loc[overall_ratio.groupby('payment_type')['overall_ratio'].idxmax()]
```

```python
# Raggruppa per tipo di pagamento e clustered time, calcolando la somma delle mance e delle corse
overall_ratio = df.groupby(['payment_type', 'clustered_pu_time']).agg(
    total_tip_amount=('tip_amount', 'sum'),
    total_fare_amount=('fare_amount', 'sum')
).reset_index()

# Calcola il rapporto tra la somma delle mance e la somma degli importi delle corse
overall_ratio['overall_ratio'] = overall_ratio['total_tip_amount'] / overall_ratio['total_fare_amount']

# Filtra le righe con valori non nulli per il rapporto
overall_ratio = overall_ratio[overall_ratio['overall_ratio'].notnull()]

# Trova il massimo rapporto per ogni tipo di pagamento
max_overall_ratio = overall_ratio.loc[overall_ratio.groupby('payment_type')['overall_ratio'].idxmax()]

# Visualizza il risultato
max_overall_ratio
```

## 17.Build chains of trips. Two trips are consecutive in a chain if (a) they have the same VendorID, (b) the pickup location of the second trip is also the dropoff location of the first trip, (c) the pickup time of the second trip is after the dropoff time of the first trip, and (d) the pickup time of the second trip is at most 2 minutes later than the dropoff time of the first trip.


Hint: Add a column chain to the dataset. A chain can have more than two trips.

```python
new=df.sort_values(['VendorID','tpep_pickup_datetime','tpep_dropoff_datetime']).reset_index(drop=True)
```

```python
new.head(20)
```

```python
new['chain'] = None
piccolo=new.head(50000)
```

```python
from datetime import timedelta
```

```python
# Inizializzazione della variabile 'chain' e del dizionario per memorizzare i dati
location_info = {}

# Assegna il valore 1 alla prima corsa
piccolo.loc[0, 'chain'] = 1
location_info[1] = {
    'tpep_dropoff_datetime': piccolo.loc[0, 'tpep_dropoff_datetime'],
    'DOLocationID': piccolo.loc[0, 'DOLocationID'],
    'VendorID': piccolo.loc[0, 'VendorID']
}

# Variabile per gestire il numero della catena
current_chain_value = 1

# Itera sulle osservazioni a partire dalla seconda
for i in range(1, len(piccolo)):
    pickup_time = piccolo.loc[i, 'tpep_pickup_datetime']
    pickup_location = piccolo.loc[i, 'PULocationID']
    current_vendor = piccolo.loc[i, 'VendorID']
    
    # Flag per indicare se è stata trovata una chain esistente
    chain_found = False
    
    # Controlla tutte le chain esistenti
    for chain_id, info in location_info.items():
        dropoff_time = info['tpep_dropoff_datetime']
        dropoff_location = info['DOLocationID']
        vendor = info['VendorID']

        # Verifica se la corsa attuale si collega a questa chain senza sovrapposizioni
        if (
            (timedelta(0) < (pickup_time - dropoff_time) <= timedelta(minutes=2)) and
            (pickup_location == dropoff_location) and
            (current_vendor == vendor)
        ):
            # Assegna alla corsa attuale la stessa chain_id
            piccolo.loc[i, 'chain'] = chain_id
            # Aggiorna le informazioni della chain
            location_info[chain_id] = {
                'tpep_dropoff_datetime': piccolo.loc[i, 'tpep_dropoff_datetime'],
                'DOLocationID': piccolo.loc[i, 'DOLocationID'],
                'VendorID': piccolo.loc[i, 'VendorID']
            }
            chain_found = True
            break
    
    # Se non è stata trovata nessuna chain compatibile, crea una nuova chain
    if not chain_found:
        current_chain_value += 1
        piccolo.loc[i, 'chain'] = current_chain_value
        location_info[current_chain_value] = {
            'tpep_dropoff_datetime': piccolo.loc[i, 'tpep_dropoff_datetime'],
            'DOLocationID': piccolo.loc[i, 'DOLocationID'],
            'VendorID': piccolo.loc[i, 'VendorID']
        }

# Il dataframe 'piccolo' ora ha la colonna 'chain' popolata correttamente senza sovrapposizioni temporali
```
